```python
import json, random

# Load raw dataset
with open("../data/dataset_raw.jsonl") as f:
    data = [json.loads(line) for line in f]

# Shuffle
random.shuffle(data)

# Split
n = len(data)
train = data[:int(0.8*n)]
val   = data[int(0.8*n):int(0.9*n)]
test  = data[int(0.9*n):]

# Write function
def write_jsonl(path, items):
    with open(path, "w") as f:
        for item in items:
            f.write(json.dumps(item, ensure_ascii=False) + "\n")

# Save files
write_jsonl("../data/train.jsonl", train)
write_jsonl("../data/val.jsonl", val)
write_jsonl("../data/test.jsonl", test)

print("Done! Train/Val/Test created.")
```
